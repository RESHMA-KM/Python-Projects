### Import necessary libraries

```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import glob
from sklearn.utils import shuffle
from skimage.transform import resize
```

### Read all the file names in all 9 folders (Daisy,Dandelion,Tulip,sunflower,orchid,Lotus,lilly,lavender,Rose)

```python
Sunflower= glob.glob('Sunflower/*jpg')
Rose = glob.glob('Rose/*.jpg')
Lavender = glob.glob('Lavender/*.jpg')
Lilly = glob.glob('Lilly/*.jpg')
Lotus = glob.glob('Lotus/*.jpg')
```

```python
len(Lavender)
```

### Read image data from all files and store it in a variable, Store the label as well

```python
image = []
label = []
    
for i in Lavender:
    img = plt.imread(i)
    image.append(img)
    label.append(0)
    
for i in Lilly:
    img = plt.imread(i)
    image.append(img)
    label.append(1)

for i in Lotus:
    img = plt.imread(i)
    image.append(img)
    label.append(2)
    
for i in Rose:
    img = plt.imread(i)
    image.append(img)
    label.append(3)
    
for i in Sunflower:
    img = plt.imread(i)
    image.append(img)
    label.append(4)
```

### Plot some of the images

```python
plt.figure(figsize = (10,8))
for i in range(12):
    plt.subplot(3,4,i+1)
    plt.imshow(image[i])
plt.tight_layout()
```

### Shuffle images and labels


```python
image, label = shuffle(image, label)
```

### Plot images once again

```python
plt.figure(figsize = (10,8))
for i in range(12):
    plt.subplot(3,4,i+1)
    plt.imshow(image[i])
plt.tight_layout()
```

### Check the shape of some of the images

```python
for i in range(10):
    print(image[i].shape)
```

### Resize all the images to a same height and width

```python
X = []

for i in image:
    X.append(resize(i, (500,500,3)))
```

 ### Convert images and labels to numpy array and store them as X and y variables

```python
X = np.array(X)
```

```python
y = np.array(label)
```

### Reshape the images to 1 dimension

```python
d1, d2, d3, d4 = X.shape
```

```python
X = X.reshape((d1, d2*d3*d4))
```

### Train test Split

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(X,y, test_size = 0.2)
```

### Create SVM Model and train it

```python
from sklearn.svm import SVC
```

```python
model = SVC()
```

```python
time model.fit(X_train, y_train)
```

### Make predictions with X_test

```python
%time y_pred = model.predict(X_test)
```

### Evaluate the Model

```python
from sklearn.metrics import accuracy_score, confusion_matrix, classification_report
```

```python
accuracy_score(y_test,y_pred)
```

```python
sns.heatmap(confusion_matrix(y_test,y_pred), annot = True)
```

```python
print(classification_report(y_test,y_pred))
```
